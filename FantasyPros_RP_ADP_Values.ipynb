```
# sort by saves, holds, SO/W, SO9
```

```
so_leaders = df2.sort_values(by=['SO9'])
```

```
so_leaders
```

```
so_leaders = so_leaders.loc[so_leaders['IP'] >= 25.0]
```

```
so_leaders = so_leaders.sort_values(by=['SO9'], ascending=False)
```

```
so_leaders.head(30)
```

```
# name, so9, so/w

name = so_leaders['Name']
team_2 = so_leaders['Tm']
so9 = so_leaders['SO9']
so_w = so_leaders['SO/W']
```

```
so_kings = pd.concat([name, team_2, so9, so_w], join = 'outer', axis = 1)
```

```
so_kings.head(15)
```

```
so_kings = so_kings.sort_values(by=['SO/W'], ascending=False)
so_kings.head(15)
```
